## Visual Analysis


```python
custom_palette = ["#FBDB93", "#BE5B50", "#8A2D3B", "#641B2E"]
sns.set_palette(custom_palette)
```

#### Age Distibution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True, color="#FBDB93")
plt.title("Age Distribution", color="#641B2E")
plt.xlabel("Age", color="#641B2E")
plt.ylabel("Count", color="#641B2E")
plt.show()
```

#### Passenger Count Per Class

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df, palette=custom_palette)
plt.title("Passenger Class Count", color="#641B2E")
plt.show()
```

#### Survival Count of Male And Female

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df, palette=custom_palette)
plt.title("Survival Count by Gender", color="#641B2E")
plt.show()
```

#### Survival Chance By Passenger Class

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df, palette=custom_palette, ci=None)
plt.title("Survival Rate by Passenger Class", color="#641B2E")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Fare', data=df, palette=custom_palette)
plt.title("Fare Distribution Among Survived & Not Survived", color="#641B2E")
plt.show()
```

#### Family Size Impact on survival

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
plt.figure(figsize=(8,5))
sns.barplot(x='FamilySize', y='Survived', data=df, palette=custom_palette, ci=None)
plt.title("Survival Rate by Family Size", color="#641B2E")
plt.show()
```

#### Age Impact On Survival

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df[df['Survived'] == 1]['Age'], color="#FBDB93", label="Survived", shade=True)
sns.kdeplot(df[df['Survived'] == 0]['Age'], color="#BE5B50", label="Not Survived", shade=True)
plt.title("Age Distribution Among Survived & Not Survived", color="#641B2E")
plt.legend()
plt.show()
```

#### Correlation Among the Numerical value

```python
data_num=df.select_dtypes(include=["number"])
plt.figure(figsize=(10,6))
sns.heatmap(data_num.corr(), annot=True, cmap=["#FBDB93", "#BE5B50", "#8A2D3B", "#641B2E"], fmt=".2f")
plt.title("Feature Correlation Heatmap", color="#641B2E")
plt.show()
```
